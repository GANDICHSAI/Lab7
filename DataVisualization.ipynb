```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sbn
```

```python
movie_data = pd.read_csv(r"IMDB Top 250 Movies.csv")
```

```python
sbn.scatterplot(data=movie_data[0:10], x='rating', y='box_office')
plt.title('Rating vs. Box Office')
```

```python
top_5_movies = movie_data.sort_values(by='rating', ascending=False).head(5)

plt.figure(figsize=(10, 6))
sbn.barplot(data=top_5_movies, x='name', y='rating', palette='viridis')
plt.title('Top 5 Movies by Rating')
```

```python

sbn.histplot(movie_data['rating'], bins=10)
plt.title('Distribution of Ratings')
plt.xlabel('Movie Rating')
plt.ylabel('Movie Frequency')
```

```python

sbn.lineplot(data=movie_data[0:5], x='genre', y='run_time')

plt.title('Movie Run Times by Genre')

```

```python
director_counts = movie_data['directors'].value_counts()
top_10_directors = director_counts.head(10)
top_10_movies = movie_data[movie_data['directors'].isin(top_10_directors.index)]


plt.figure(figsize=(10, 6))
sbn.countplot(y='directors', data=top_10_movies, order=top_10_directors.index, palette='viridis')
plt.xlabel('Number of Movies Directed')
plt.ylabel('Director')
plt.title('Top 10 Directors and Their Movies')
plt.show()
```

```python

cleaned_movie_data = movie_data.dropna()
cleaned_movie_data.plot.scatter(x ='year', y ='rank', c ='rating', cmap ='coolwarm') 
```
